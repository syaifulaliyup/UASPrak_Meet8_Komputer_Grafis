Distribusi data pada atribut Puechase Price

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('DataPrak8Final.csv')

print(df.head())

# Mengonversi kolom 'Purchase Price' ke tipe data numerik
df['Purchase Price'] = df['Purchase Price'].replace('[\$,]', '', regex=True).astype(float)

# Mengatur ukuran grafik
plt.figure(figsize=(10, 6))

# Membuat histogram menggunakan seaborn
sns.histplot(df['Purchase Price'], bins=10, kde=True, color='#AF47D2', edgecolor='black')
plt.title('Distribusi Purchase Price dari Data CSV')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')

plt.show()
```


b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('DataPrak8Final.csv')

# Mengambil 5 saham tertinggi
top_5_shares = df.nlargest(5, 'Shares')

# Mengatur ukuran grafik
plt.figure(figsize=(12, 8))

# Membuat grafik menggunakan seaborn
sns.barplot(x='Shares', y='Stock Name', hue='Symbol', data=top_5_shares, palette='Accent')

# Menambahkan judul dan label
plt.title('5 Saham dengan Jumlah Shares Tertinggi')
plt.xlabel('Shares')
plt.ylabel('Stock Name')

plt.show()
```

c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
df = pd.read_csv('DataPrak8Final.csv')

# Mengonversi kolom 'Market Value' ke tipe data numerik, menghilangkan tanda dolar dan koma
df['Market Value'] = df['Market Value'].replace('[\$,]', '', regex=True).replace(',', '', regex=True).astype(float)

# Mengambil 5 nilai tertinggi berdasarkan Market Value
top_5_market_value = df.nlargest(5, 'Market Value')

custom_palette = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#33FFA6']

# Mengatur ukuran grafik
plt.figure(figsize=(12, 8))

# Membuat grafik menggunakan seaborn
sns.barplot(x='Market Value', y='Stock Name', hue='Symbol', data=top_5_market_value, palette=custom_palette)

# Menambahkan judul dan label
plt.title('5 Saham dengan Market Value Tertinggi')
plt.xlabel('Market Value')
plt.ylabel('Stock Name')

# Menampilkan grafik
plt.show()
```

d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('DataPrak8Final.csv')

# Mengonversi kolom 'Annual Yield' ke tipe data numerik (menghilangkan karakter persen % dan mengubah ke float)
df[' Annual Yield'] = df[' Annual Yield'].replace('%', '', regex=True).astype(float) / 100.0

# Mengambil 7 nilai tertinggi berdasarkan Annual Yield
top_7_annual_yield = df.nlargest(7, ' Annual Yield')

# Mengatur ukuran grafik
plt.figure(figsize=(12, 8))

# Membuat grafik menggunakan seaborn
sns.barplot(x=' Annual Yield', y='Stock Name', hue='Symbol', data=top_7_annual_yield, palette='Paired')

# Menambahkan judul dan label
plt.title('7 Saham dengan Annual Yield Tertinggi')
plt.xlabel(' Annual Yield')
plt.ylabel('Stock Name')

plt.show()
```
